```python
# Scanner without LEX

keywords = {
    "int", "float", "include", "if", "while", "return"
}

operators = {
    "+", "-", "*", "/"
}

seperators = {
    ";", ",", "(", ")", "{", "}"
}

with open("input.text", "r") as file:
  source = file.read()

word = ""
number = ""

i = 0

while i < len(source):
  ch = source[i]

  if ch.isalpha():
    word += ch
  elif ch.isdigit():
    number += ch
  elif ch == '"':
    str_var = ""
    i += 1
    while i < len(source) and source[i] != '"':
      str_var += source[i]
      i += 1
    i += 1
    print("<", str_var, ", STRING>")
  else:
    if word:
      if word in keywords:
        print("<", word, ", KEYWORD>")
      else:
        print("<", word, ", IDENTIFIER>")
      word = ""
    elif number:
      print("<", number, ", NUMBER>")
      number = ""
    elif ch in operators:
      print("<", ch, ", OPERATOR>")
    elif ch in seperators:
      print("<", ch, ", SEPERATOR>")
  i += 1

if word:
  if word in keywords:
    print("<", word, ", KEYWORD>")
  else:
    print("<", word, ", IDENTIFIER>")
elif number:
  print("<", number, ", NUMBER>")
```

```python
# DFA
n = int(input("Number of states : "))

transitions = {}

for state in range(n):
  print("State", state)
  transitions[state] = {}

  transitions[state]["0"] = int(input("On 0 goes to : "))
  transitions[state]["1"] = int(input("On 1 goes to : "))

initial_state = int(input("Initial state :"))

final_states = list(map(int, input("Enter final states : ").split()))

input_str = input("Enter string : ")

current_state = initial_state

for symbol in input_str:
  current_state = transitions[current_state][symbol]

if current_state in final_states:
  print("Accepted")
else:
  print("Rejected")
```

```python
# RDP

grammar = {}

n = int(input("Enter number of productions : "))

for _ in range(n):
  prod = input("Enter production : ").split("->")
  lhs = prod[0].strip()
  rhs = prod[1].strip().split("|")
  grammar[lhs] = rhs

start_symbol = list(grammar.keys())[0]

# Fix: Convert input string directly to a list of characters
input_str = list(input("Enter input string : "))

def parse(symbol, pos):
  # Terminal
  if symbol not in grammar:
    if pos < len(input_str) and input_str[pos] == symbol:
      return pos + 1
    else:
      return -1
  # Non terminal
  for production in grammar[symbol]:
    cur_pos = pos
    valid = True

    for ch in production:
      next_pos = parse(ch, cur_pos)
      if next_pos == -1:
        valid = False
        break
      cur_pos = next_pos
    if valid:
      return cur_pos
  return -1

final_pos = parse(start_symbol, 0)

if final_pos == len(input_str):
  print("String Accepted")
else:
  print("String rejected")
```

```python
# Optimization techniques

# Code Optimization Techniques

# 1. Constant Folding
def constant_folding(code):

    result = []

    for line in code:

        lhs, rhs = line.split("=")

        parts = rhs.split()

        if len(parts) == 3:

            op1, op, op2 = parts

            if op1.isdigit() and op2.isdigit():

                if op == "+":
                    rhs = str(int(op1) + int(op2))

                elif op == "-":
                    rhs = str(int(op1) - int(op2))

                elif op == "*":
                    rhs = str(int(op1) * int(op2))

                elif op == "/":
                    rhs = str(int(op1) // int(op2))

        result.append(lhs + "=" + rhs)

    return result


# 2. Constant Propagation
def constant_propagation(code):

    const = {}
    result = []

    for line in code:

        lhs, rhs = line.split("=")

        for var in const:
            rhs = rhs.replace(var, const[var])

        if rhs.isdigit():
            const[lhs] = rhs

        result.append(lhs + "=" + rhs)

    return result


# 3. Common Subexpression Elimination
def common_subexpr(code):

    expr = {}
    result = []

    for line in code:

        lhs, rhs = line.split("=")

        if rhs in expr:
            result.append(lhs + "=" + expr[rhs])

        else:
            expr[rhs] = lhs
            result.append(line)

    return result


# 4. Loop Invariant Code Motion
def loop_invariant(code):

    outside = []
    inside = []

    for line in code:

        if "i" in line:
            inside.append(line)
        else:
            outside.append(line)

    return outside + inside


# 5. Dead Code Elimination
def dead_code(code):

    used = set()
    result = []

    for line in reversed(code):

        if "=" in line:

            lhs, rhs = line.split("=")

            lhs = lhs.strip()
            rhs = rhs.strip()

            if lhs in used or len(result) == 0:

                result.append(line)

                for ch in rhs:

                    if ch.isalpha():
                        used.add(ch)

    result.reverse()
    return result


# Input

n = int(input("Enter number of statements: "))

code = []

for i in range(n):
    code.append(input())

print("\nConstant Folding")
print(constant_folding(code))

print("\nConstant Propagation")
print(constant_propagation(code))

print("\nCommon Subexpression Elimination")
print(common_subexpr(code))

print("\nLoop Invariant Code Motion")
print(loop_invariant(code))

print("\nDead Code Elimination")
print(dead_code(code))
```

```python
# TAC

# Three Address Code Generation

def precedence(op):

    if op in ['+', '-']:
        return 1

    if op in ['*', '/']:
        return 2

    return 0


# Infix to Postfix
def infix_to_postfix(expr):

    stack = []
    postfix = []

    for ch in expr:

        if ch.isalnum():
            postfix.append(ch)

        elif ch == '(':
            stack.append(ch)

        elif ch == ')':

            while stack and stack[-1] != '(':
                postfix.append(stack.pop())

            stack.pop()

        else:

            while (stack and
                   precedence(stack[-1]) >= precedence(ch)):
                postfix.append(stack.pop())

            stack.append(ch)

    while stack:
        postfix.append(stack.pop())

    return ''.join(postfix)


# Generate TAC
def generate_tac(postfix):

    stack = []
    temp_count = 1

    print("\nThree Address Code:\n")

    for ch in postfix:

        if ch.isalnum():
            stack.append(ch)

        else:

            op2 = stack.pop()
            op1 = stack.pop()

            temp = "t" + str(temp_count)
            temp_count += 1

            print(temp, "=", op1, ch, op2)

            stack.append(temp)

    return stack.pop()


# Input

expr = input("Enter Expression: ")

lhs, rhs = expr.split("=")

postfix = infix_to_postfix(rhs)

result = generate_tac(postfix)

print(lhs, "=", result)
```

```python
# Assembly Code Generation from TAC

def generate_assembly(tac_lines):

    register_number = 1
    temporary_to_register = {}

    print("\nAssembly Code:\n")

    for statement in tac_lines:

        tokens = statement.split()

        # Example: t1 = c * d
        if len(tokens) == 5:

            result = tokens[0]
            left_operand = tokens[2]
            operator = tokens[3]
            right_operand = tokens[4]

            current_register = "R" + str(register_number)
            register_number += 1

            left_value = temporary_to_register.get(
                left_operand,
                left_operand
            )

            right_value = temporary_to_register.get(
                right_operand,
                right_operand
            )

            print(f"MOV {current_register}, {left_value}")

            if operator == "+":
                print(f"ADD {current_register}, {right_value}")

            elif operator == "-":
                print(f"SUB {current_register}, {right_value}")

            elif operator == "*":
                print(f"MUL {current_register}, {right_value}")

            elif operator == "/":
                print(f"DIV {current_register}, {right_value}")

            temporary_to_register[result] = current_register

        # Example: a = t2
        elif len(tokens) == 3:

            destination = tokens[0]
            source = tokens[2]

            source_register = temporary_to_register.get(
                source,
                source
            )

            print(f"MOV {destination}, {source_register}")


tac = []

print("Enter TAC Statements (Type END to stop):")

while True:

    line = input()

    if line.upper() == "END":
        break

    tac.append(line)

generate_assembly(tac)
```

```python
# LL(1)

# LL(1) Parser - FIRST, FOLLOW, PARSE TABLE

grammar = {}
first_set = {}
follow_set = {}

# Input Grammar
n = int(input("Enter number of productions: "))

for i in range(n):

    lhs, rhs = input().replace(" ", "").split("->")

    grammar[lhs] = rhs.split("|")

start = list(grammar.keys())[0]


# FIRST
def first(symbol):

    # Terminal
    if not symbol.isupper():
        return {symbol}

    # Already computed
    if symbol in first_set:
        return first_set[symbol]

    result = set()

    for production in grammar[symbol]:

        if production == 'ε':

            result.add('ε')

        else:

            for ch in production:

                temp = first(ch)

                result |= (temp - {'ε'})

                if 'ε' not in temp:
                    break

            else:
                result.add('ε')

    first_set[symbol] = result

    return result


# FOLLOW
def follow(symbol):

    if symbol in follow_set:
        return follow_set[symbol]

    result = set()

    # Start symbol gets $
    if symbol == start:
        result.add('$')

    for lhs in grammar:

        for production in grammar[lhs]:

            for i in range(len(production)):

                if production[i] == symbol:

                    # Symbol followed by something
                    if i + 1 < len(production):

                        next_symbol = production[i + 1]

                        temp = first(next_symbol)

                        result |= (temp - {'ε'})

                        if 'ε' in temp and lhs != symbol:
                            result |= follow(lhs)

                    # Symbol at end
                    else:

                        if lhs != symbol:
                            result |= follow(lhs)

    follow_set[symbol] = result

    return result


# Compute FIRST
for nt in grammar:
    first(nt)

# Compute FOLLOW
for nt in grammar:
    follow(nt)


# Print FIRST
print("\nFIRST")

for nt in grammar:
    print(nt, "=", first_set[nt])

# Print FOLLOW
print("\nFOLLOW")

for nt in grammar:
    print(nt, "=", follow_set[nt])


# PARSE TABLE

table = {}

for nt in grammar:
    table[nt] = {}

for lhs in grammar:

    for production in grammar[lhs]:

        prod_first = set()

        # FIRST of production
        if production == 'ε':

            prod_first.add('ε')

        else:

            for ch in production:

                temp = first(ch)

                prod_first |= (temp - {'ε'})

                if 'ε' not in temp:
                    break

            else:
                prod_first.add('ε')

        # Fill FIRST entries
        for terminal in prod_first - {'ε'}:

            table[lhs][terminal] = lhs + "->" + production

        # Fill FOLLOW entries if epsilon exists
        if 'ε' in prod_first:

            for terminal in follow_set[lhs]:

                table[lhs][terminal] = lhs + "->" + production


# Print Parse Table

print("\nPARSE TABLE")

for nt in table:
    print(nt, ":", table[nt])
```
